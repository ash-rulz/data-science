```python
import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
df.head()
```

# Scatter Plot

```python
import seaborn as sns

sns.set()
sns.scatterplot(x="# Gears", y="CombMPG", data = df)
```

# Joinplot

Joinplot contains both scatterplot and histogram in the same diagram. It gives the frequency of each axis element.

```python
sns.jointplot(x='# Gears', y='CombMPG', data = df)
```

# lmplot

lmplot is scatterplot with liner regression line computed.

```python
sns.lmplot(x="# Gears", y="CombMPG", data = df)
```

# Boxplot/Box and Whiskers plot

Useful for visualizating typyical values for a given category. Each box represents the range between the 1st and the 3rd quartiles.

```python
sns.boxplot(x='# Gears', y='CombMPG', data = df)
```

# Swarmplot

Swarmplot is just like boxplot, but witout the boxes.

```python
sns.swarmplot(x= "# Gears", y="CombMPG", data = df)
```
