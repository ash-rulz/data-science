Transpose the columns to rows - use melt()

```python
import pandas as pd
pew = pd.read_csv('pew.csv')
pew.head()
```

```python
pew.info()
```

```python
pew.melt(id_vars='religion')
```

```python
pew_tidy = pew.melt(id_vars = 'religion', var_name = 'incomeGrp', value_name = 'count')
pew_tidy.head()
```

```python
billboard = pd.read_csv('billboard.csv')
billboard.info()
```

```python
billboard.head()
```

```python
(billboard.melt(id_vars = ['year', 'artist', 'track', 'time', 'date.entered'], var_name = 'week',
                  value_name = 'rank').groupby('artist').agg(avg_rnk = ('rank', 'mean'))
                .sort_values(by='avg_rnk', ascending = True)#Keep doing operations by using dot
)#use the bracket for multi-line
```

```python
ebola = pd.read_csv('country_timeseries.csv')
ebola.head()
```

```python
ebola_long = ebola.melt(id_vars=['Date', 'Day'], var_name ='cd_Country', value_name = 'Count')
ebola_long.head()
```

```python
ebola_split=ebola_long.cd_Country.str.split('_',expand=True)
ebola_split.head()
```

```python
ebola_long[['Status','Country']] = ebola_split
ebola_long.head()
```

```python
weather = pd.read_csv('weather.csv')
weather.head()
```

```python
weather_long = weather.melt(id_vars=['id', 'year', 'month', 'element'], 
                            var_name = 'day', value_name = 'temp')
weather_long.head()
```

```python
weather_long.day.unique()
```

```python
#To pivot rows to columns, use the pivot_table function
(
    weather_long.pivot_table(index=['id','year','month','day'], 
                             columns='element',
                             values='temp').reset_index()
)
```
