```python
df.groupby(['year'])['lifeExp'].mean()
```

```python
import numpy as np
df.groupby(['year'])['lifeExp'].agg(np.mean)#Gives the same result  as above
```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].agg(np.mean)
```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].agg(np.mean).reset_index()#To flatten it out
```
